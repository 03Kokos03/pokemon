```python
data = [[poke_mean1, poke_mean_HP1, 1], 
        [poke_mean2, poke_mean_HP2, 2],
        [poke_mean3, poke_mean_HP3, 3] ,
        [poke_mean6, poke_mean_HP4, 4], 
        [poke_mean5, poke_mean_HP5, 5], 
        [poke_mean6, poke_mean_HP6, 6]]
poke_mean = pd.DataFramedf = pd.DataFrame(data, columns=['mean_total', 'mean_HP', 'generation'])

poke_mean




```

```python

poke_mean.plot(x ='generation' , y = 'mean_total' , label='mean_total' , linestyle="--" ,marker='.' , markersize=8)
plt.xticks([1,2,3,4,5,6])
plt.yticks([400, 410, 420, 430, 440 ,450])
plt.savefig('graph_poke_2.png')

poke_mean.plot(x ='generation' , y = 'mean_HP' , label='mean_HP' , linestyle="--" ,marker='.' , markersize=8 , color='red')
plt.xticks([1,2,3,4,5,6])
plt.yticks([55, 60, 65, 70, 75 ,80])

plt.legend()

plt.savefig('graph_poke_1.png')
plt.show()
```
